```python
visa_df.iloc[<rows>,<columns>]   #[start,end]
visa_df.iloc[<start:end>,<start:end>]
visa_df.iloc[[rows index],[columns index]]
```

```python
visa_df.iloc[0:4,0:5]
#i want to select rows has index 0 to 4
# i want to select columns has index 0 to 5
```

```python
visa_df.iloc[5:10,]
visa_df.iloc[5:10, :] #same will come
```

```python
visa_df.iloc[:,: 2]
```

```python
visa_df.iloc[5:9,2:6]# 5 to 9 index rows, 2 to 5 index columns

visa_df.iloc[5:10,] #all columns, 5 to 9 index rows

visa_df.iloc[:,: 2] #all rows, only 0 and 1 colums
```

```python
rows=[100,200,300] #specific rows & columns #random samples
cols=[2,7]
visa_df.iloc[rows,cols]
```

```python
rows=[100,200,300] 
cols=['education_of_employee','region_of_employment']
visa_df.iloc[rows,cols]

#fails iloc requires numeric indexers
```

```python
rows=[100,200,300] 
cols=['education_of_employee','region_of_employment']
visa_df.loc[rows,cols] #remove the i from iloc
```
